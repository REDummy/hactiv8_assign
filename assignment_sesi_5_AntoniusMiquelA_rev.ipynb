```python
menu = {
    "Espresso": 25000,
    "Latte": 35000,
    "Cappuccino": 38000,
    "Croissant": 28000,
    "Sandwich": 42000,
    "Cheesecake": 45000,
}

transaksi = [
    {"nama": "Budi", "pesanan": ["Latte", "Croissant", "Croissant"]},
    {"nama": "Sari", "pesanan": ["Espresso", "Cheesecake"]},
    {"nama": "Dian", "pesanan": ["Cappuccino", "Sandwich", "Latte", "Cheesecake"]},
    {"nama": "Reza", "pesanan": ["Espresso"]},
    {"nama": "Ayu",  "pesanan": ["Latte", "Latte", "Sandwich", "Croissant", "Cheesecake"]},
]
```

```python
# task 1
total_pendapatan = 0
tagihan = {}

for i in transaksi :
    nama = i["nama"]
    pesanan = i["pesanan"]

    subtotal = 0
    for item in pesanan :
        subtotal += menu[item]

    if subtotal <50000 : 
        diskon_pct = 0
    elif subtotal < 100000 :
        diskon_pct = 10
    else : 
        diskon_pct = 15

    diskon_rp = subtotal * diskon_pct/100
    total = subtotal - diskon_rp

    tagihan[nama] = total

    total_pendapatan += total

    print(f"Pelanggan : {nama}")
    print(f"Pesanan : {', '.join(pesanan)}")
    print(f"Subtotal : Rp. {subtotal:,}".replace(',','.'))
    print(f"Diskon   : {diskon_pct}% (Rp {int(diskon_rp):,})".replace(",", "."))
    print(f"Total    : Rp {int(total):,}".replace(",", "."))
    print("-" * 67) #ayeeee
    print(f"\nTotal Pendapatan Kafe: Rp {int(total_pendapatan):,}".replace(",", "."))

  
```

```python
# task 2

pelanggan_max = None
nilai_max = 0

for nama, total in tagihan.items() :
    if total > nilai_max:
        nilai_max = total
        pelanggan_max = nama

print(f"Pelanggan dengan tagihan terbesar :{pelanggan_max}")
print(f"Total tagihan: Rp {int(nilai_max):,}".replace(",", "."))
```

```python
# TASK 3

antrian = ["Budi", "Sari", "Dian", "Reza", "Ayu"]
stok_kopi = 3
kopi_menu = ["Espresso", "Latte", "Cappuccino"]

data_pesanan = {t["nama"]: t["pesanan"] for t in transaksi}

while len(antrian) > 0 and stok_kopi > 0 : 
    cust = antrian.pop(0)
    order = data_pesanan[cust]

    jumlah_kopi = 0
    for i in order : 
        if i in kopi_menu:
            jumlah_kopi += 1

    stok_kopi -= jumlah_kopi
    if stok_kopi < 0 : stok_kopi = 0 

    print(f"Melayani: {cust} | Stok kopi tersisa: {stok_kopi}")

    if stok_kopi == 0:
        print(f"Antrian berhenti di : {cust}")
        break

if stok_kopi == 0 and len(antrian) > 0:
    print("Stok kopi habis. Pesanan berikutnya tidak dapat dilayani.")
elif len(antrian) == 0:
    print("Semua pelanggan telah dilayani.")
```
