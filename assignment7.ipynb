```python
# TASK 1

class Karyawan:
    def __init__(self, nama, departemen, gaji_pokok):
        self.nama = nama
        self.departemen = departemen
        self.__gaji_pokok = gaji_pokok  

    def get_gaji_pokok(self):
        return self.__gaji_pokok

    def set_gaji_pokok(self, nilai):
        if nilai > 0:
            self.__gaji_pokok = nilai
        else:
            print("Error: Gaji pokok harus lebih besar dari 0.")

    def hitung_gaji(self):
        return self.__gaji_pokok

    def info(self):
        print(f"Nama: {self.nama}")
        print(f"Departemen: {self.departemen}")
        print(f"Gaji Pokok: {self.__gaji_pokok}")
        print(f"Total Gaji: {self.hitung_gaji()}")
        print("-" * 67)


# Uji Task 1
k1 = Karyawan("Andi", "IT", 5000000)
k1.info()
```

```python
# TASK 2
class KaryawanTetap(Karyawan):
    def __init__(self, nama, departemen, gaji_pokok, tunjangan_tetap, bonus_tahunan):
        super().__init__(nama, departemen, gaji_pokok)
        self.tunjangan_tetap = tunjangan_tetap
        self.bonus_tahunan = bonus_tahunan

    def hitung_gaji(self):
        total = self.get_gaji_pokok() + self.tunjangan_tetap + (self.bonus_tahunan / 12)
        return int(total)

    def info(self):
        print("Jenis: Karyawan Tetap")
        print(f"Nama: {self.nama}")
        print(f"Departemen: {self.departemen}")
        print(f"Gaji Pokok: {self.get_gaji_pokok()}")
        print(f"Tunjangan: {self.tunjangan_tetap}")
        print(f"Bonus per bulan: {int(self.bonus_tahunan / 12)}")
        print(f"Total Gaji: {self.hitung_gaji()}")
        print("-" * 67)


# Uji Task 2
kt1 = KaryawanTetap("Budi", "Finance", 7000000, 1500000, 12000000)
kt2 = KaryawanTetap("Siti", "HR", 6500000, 1200000, 6000000)

kt1.info()
kt2.info()
```

```python
# TASK 3

class KaryawanKontrak(Karyawan):
    def __init__(self, nama, departemen, gaji_pokok, durasi_kontrak_bulan, rate_per_jam, jam_per_bulan):
        super().__init__(nama, departemen, gaji_pokok)
        self.durasi_kontrak_bulan = durasi_kontrak_bulan
        self.rate_per_jam = rate_per_jam
        self.jam_per_bulan = jam_per_bulan

    def hitung_gaji(self):
        return self.rate_per_jam * self.jam_per_bulan

    def info(self):
        print("Jenis: Karyawan Kontrak")
        print(f"Nama: {self.nama}")
        print(f"Departemen: {self.departemen}")
        print(f"Durasi Kontrak: {self.durasi_kontrak_bulan} bulan")
        print(f"Rate per jam: {self.rate_per_jam}")
        print(f"Jam per bulan: {self.jam_per_bulan}")
        print(f"Total Gaji: {self.hitung_gaji()}")
        print("-" * 67)


# Uji Task 3
kk1 = KaryawanKontrak("Rina", "IT", 0, 12, 50000, 160)
kk2 = KaryawanKontrak("Dodi", "Design", 0, 6, 60000, 150)

kk1.info()
kk2.info()
```

```python
# List semua karyawan (total 4)
tim_techspark = [kt1, kt2, kk1, kk2]

total_gaji = 0

for karyawan in tim_techspark:
    nama = karyawan.nama
    jenis = type(karyawan).__name__
    gaji = karyawan.hitung_gaji()

    print(f"Nama: {nama}, Tipe: {jenis}, Gaji: {gaji}")
    total_gaji += gaji

print("\nTotal Pengeluaran Gaji per bulan:", total_gaji)
print("=" * 67)
```
