```python
produk = [
    {'nama': 'Laptop ASUS A516', 'harga': 7_500_000, 'stok': 12, 'kategori': 'laptop'},
    {'nama': 'Mouse Logitech M100', 'harga': 85_000, 'stok': 50, 'kategori': 'aksesoris'},
    {'nama': 'Keyboard Mechanical Rexus', 'harga': 450_000, 'stok': 30, 'kategori': 'aksesoris'},
    {'nama': 'Monitor LG 24 inch', 'harga': 2_300_000, 'stok': 8, 'kategori': 'monitor'},
    {'nama': 'SSD Kingston 512GB', 'harga': 650_000, 'stok': 25, 'kategori': 'storage'},
]
```

```python
# Task 1

for p in produk :
    kategori = p['kategori'].upper()
    harga =  f"{p['harga']:,}".replace(',','.')
    print(f'[{kategori}] {p['nama']} | Harga: Rp {harga} | Stok: {p['stok']} unit')
```

```python
# Task 2
total_nilai_inv = 0
total_stok = 0
for p in produk : 
    total_nilai_inv += p["harga"] * p["stok"]
    total_stok += p["stok"]


total_format = f"{total_nilai_inv:,}".replace(",", ".")
print(f"Total Nilai Inventory: Rp {total_format}")

rata_harga = total_nilai_inv / total_stok
print(f"Rata-rata Harga Produk: Rp {rata_harga:,.2f}".replace(",", "."))


```

```python
# Task 3

nama_semua_produk = " | ".join([p['nama'] for p in produk])
print(nama_semua_produk)


print("Jumlah karakter:", len(nama_semua_produk))
print("Apakah mengandung 'Rexus'?", "Rexus" in nama_semua_produk)

produk_pertama = nama_semua_produk.split(" | ")[0]
print("Produk pertama:", produk_pertama)

```

```python
# Task 4


def get_ringkasan() : 
    ringkasan_kategori = {}
    for p in produk:
        kategori = p["kategori"]
        ringkasan_kategori[kategori] = ringkasan_kategori.get(kategori, 0) + 1

    for k, v in ringkasan_kategori.items():
        print(f"Kategori [{k}]: {v} produk")


get_ringkasan()

def tambah_produk(data_produk, produk_baru, key="nama"):
    for p in data_produk:
        if p.get(key) == produk_baru.get(key):
            return  # product already exists, do nothing

    data_produk.append(produk_baru)

produk_baru = {
    'nama': 'Webcam Logitech C920', 
    'harga': 750_000, 
    'stok': 15, 
    'kategori': 'aksesoris'
}

tambah_produk(produk, produk_baru)

print("\nRingkasan Setelah di update")
get_ringkasan()

```
