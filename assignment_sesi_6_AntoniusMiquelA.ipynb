```python

penjualan = [
    {"produk": "Printer HP LaserJet",    "unit": 3,  "harga_satuan": 1800000},
    {"produk": "Tinta Epson L3150",      "unit": 10, "harga_satuan": 320000},
    {"produk": "Kertas A4 Sinar Dunia",  "unit": 25, "harga_satuan": 55000},
    {"produk": "Stapler Kenko",          "unit": 15, "harga_satuan": 35000},
    {"produk": "Meja Kantor Activ",      "unit": 2,  "harga_satuan": 2750000},
    {"produk": "Kursi Ergonomis Uno",    "unit": 4,  "harga_satuan": 1500000},
]
```

```python
def hitung_subtotal(unit, harga_satuan) : 
    return unit*harga_satuan
def hitung_diskon(subtotal, persen_diskon=0) :
    return subtotal*persen_diskon/100

def hitung_total(subtotal, persen_diskon=0) : 
    return subtotal - hitung_diskon(subtotal, persen_diskon)

for i in penjualan : 
    unit = i["unit"]
    item = i["harga_satuan"]

    subtotal = hitung_subtotal(unit, item)
    diskon = hitung_diskon(subtotal, 12)
    total = hitung_total(subtotal, 12)

    print(f"{i["produk"]}, dengan subtotal {subtotal}, diskon {diskon}, dan total akhir {total}")
```

```python
def buat_laporan(data_jual, pct_diskon=0):
    detail = []
    total_semua = 0
    total_unit = 0 


    # This was wrong, indentation error. only passing the last item instead

    for item in data_jual:
        subtotal = hitung_subtotal(item["unit"], item["harga_satuan"])
        diskon = hitung_diskon(subtotal, pct_diskon)
        total = hitung_total(subtotal, pct_diskon)

        
        total_semua += total
        total_unit += item["unit"]

        detail.append({
            **item,
            "subtotal": subtotal,
            "diskon": diskon,
            "total": total
            })
        
        rata_unit = round(total_unit/ len(data_jual),2)
        
    return {
            "detail": detail,
            "grand_total": total_semua,
            "rata_unit": rata_unit
        }


# Tanpa diskon
laporan1 = buat_laporan(penjualan)
print(laporan1)
print("Tanpa Diskon:")
print("Grand Total:", laporan1["grand_total"])
print("Rata Unit:", laporan1["rata_unit"])

# Dengan diskon 12%
laporan2 = buat_laporan(penjualan, 12)
print(laporan2)
print("\nDiskon 12%:")
print("Grand Total:", laporan2["grand_total"])
print("Rata Unit:", laporan2["rata_unit"])

        
```

```python
from datetime import datetime

def tambahkan_timestamp(laporan):
    bulan_indo = [
        "Januari","Februari","Maret","April","Mei","Juni",
        "Juli","Agustus","September","Oktober","November","Desember"
    ]

    now = datetime.now()
    formatted = f"{now.day} {bulan_indo[now.month-1]} {now.year}, {now.strftime('%H:%M:%S')}"
    
    laporan["dibuat_pada"] = formatted
    return laporan

laporan_time = tambahkan_timestamp(laporan1.copy())
print("Timestamp:", laporan_time["dibuat_pada"])
```

```python
import math

def statistik_penjualan(data_penjualan):
    subtotals = []
    total_unit = 0

    for item in data_penjualan:
        subtotal = hitung_subtotal(item["unit"], item["harga_satuan"])
        subtotals.append(subtotal)
        total_unit += item["unit"]

    mean = sum(subtotals) / len(subtotals)

    variance = sum((x - mean) ** 2 for x in subtotals) / len(subtotals)
    std_dev = math.sqrt(variance)

    return {
        "total_unit": total_unit,
        "max_subtotal": max(subtotals),
        "min_subtotal": min(subtotals),
        "std_subtotal": std_dev
    }

# Uji
stats = statistik_penjualan(penjualan)
print("\nStatistik Penjualan:")
print(stats)
```
