```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('tips')
df.head(100)

```

```python
plt.figure(figsize=(8,5))
plt.hist(df['tip'], bins=20, color='skyblue', edgecolor='black')

mean_tip = df['tip'].mean()
median_tip = df['tip'].median()

plt.axvline(mean_tip, color='red', linestyle='-.', label=f'Mean = {mean_tip:.2f}')
plt.axvline(median_tip, color='green', linestyle='-', label=f'Median = {median_tip:.2f}')

plt.title('Distribusi Tip')
plt.xlabel('Tip')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```

```python
order_days = ['Thur', 'Fri', 'Sat', 'Sun']

plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='day', y='tip', hue='sex', order=order_days)

plt.title('Distribusi Tip berdasarkan Hari')
plt.xlabel('Hari')
plt.ylabel('Tip')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x='total_bill',
    y='tip',
    hue='time',     # different colors
    style='time',   # different marker shapes
    s=80            
)

plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
df['tip_pct'] = df['tip'] / df['total_bill'] * 100

plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x='total_bill',
    y='tip_pct',
    hue='time',     # different colors
    style='time',   # different marker shapes
    s=80            
)

plt.title('Total Bill vs Tip Percentage')
plt.xlabel('Total Bill')
plt.ylabel('Tip (%)')
plt.show()
```

```python
plt.figure(figsize=(6,5))
corr = df[['total_bill', 'tip', 'size', 'tip_pct']].corr()

sns.heatmap(corr, annot=True, fmt='.3f', cmap='inferno_r')
plt.title('Heatmap Korelasi Variabel Numerik')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='sex', y='tip_pct', hue='smoker')

plt.title('Rata-rata Tip Percentage berdasarkan Sex dan Smoker')
plt.xlabel('Sex')
plt.ylabel('Tip (%)')
plt.legend(title='Smoker')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5), sharey=True)

sns.boxplot(data=df, x='smoker', y='tip', ax=axes[0])
axes[0].set_title('Tip berdasarkan Smoker')
axes[0].set_xlabel('Smoker')
axes[0].set_ylabel('Tip')

sns.boxplot(data=df, x='sex', y='tip', ax=axes[1])
axes[1].set_title('Tip berdasarkan Sex')
axes[1].set_xlabel('Sex')

fig.suptitle('Perbandingan Distribusi Tip berdasarkan Kategori Pelanggan')
plt.show()
```
